```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os

os.environ["GOOGLE_API_KEY"] = ""
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant"),
        ("human", "{input}")
    ]
)
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0)
```

```python
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()
```

## Manual Invocation

```python
template = prompt_template.invoke({"input":"What is the capital of France?"})

result = llm.invoke(template)
final_result = str_parser.invoke(result)
final_result
```

## Chain Invocation (LCEL)

LangChain Expression Language (LCEL) is a declarative way to easily compose chains together. It provides a simple syntax (using the pipe `|` operator) to connect different components like prompts, models, and output parsers, ensuring consistent interface support for features like streaming and async operations.

```python
chain = prompt_template | llm | str_parser
chain.invoke({"input":"What is the capital of France?"})
```

## RunnableSequence

`RunnableSequence` is the underlying class that LCEL uses to create a chain. It represents a sequence of runnables where the output of one step is passed as the input to the next. While the pipe syntax is usually preferred for its brevity, `RunnableSequence` provides an explicit way to construct these execution pipelines.

```python
from langchain_core.runnables import RunnableSequence

chain_1 = RunnableSequence(prompt_template, llm, str_parser)

chain_1.invoke({"input":"What is the capital of France?"})
```
